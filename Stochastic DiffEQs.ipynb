# Stchastic Differential Equation solver and visualization

```julia
using Plots, DifferentialEquations
```

## Solve Black-shoels model (Financial mathematics)

```julia
α=1
β=1
u₀=1/2
f(u,p,t) = α*u
g(u,p,t) = β*u
dt = 1//2^(4)
tspan = (0.0,1.0)
prob = SDEProblem(f,g,u₀,(0.0,1.0))
```

```julia
sol_EM = solve(prob, EM(), dt=dt)
res_EM = plot(sol_EM)
for i in 1:10
    sol_EM = solve(prob, EM(), dt=dt)
    res_EM = plot!(sol_EM)
end
```

```julia
res_EM
```

```julia
sol_SRIW = solve(prob, SRIW1(), dt=dt)
res_SRIW = plot(sol_SRIW)
for i in 1:10
    sol_SRIW = solve(prob, EM(), dt=dt)
    res_SRIW = plot!(sol_SRIW)
end
```
